# 32_50120 data:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Dataset/Span_Level/DS_SpanLev_1e4_uniform_32_50120.csv')
```

```python
df.head()
```

```python
# name of each col in df
for i in df.columns:
  print(i)
```

```python
for i in df.columns:
  if str(i) == 'Lspan(KM)'or str(i) == 'Ptx(dBm)' or str(i) == 'CutNumCh' or str(i) == 'TotCongLSP(%)':
    plt.figure()
    fig,ax = plt.subplots(figsize=(30,15))
    sns.boxplot(x = str(i), y = 'GSNRSpan(dB)',  data = df,ax=ax)
    plt.title(str(i)+' vs GSNRSpan(dB)')
    ticks = plt.setp(ax.get_xticklabels(),rotation=90)
    plt.show()
```

# Violin plot:

```python
for i in df.columns:
  if str(i) == 'Lspan(KM)'or str(i) == 'Ptx(dBm)' or str(i) == 'CutNumCh' or str(i) == 'TotCongLSP(%)':
    plt.figure()
    fig,ax = plt.subplots(figsize=(30,15))
    sns.violinplot(x = str(i), y = 'GSNRSpan(dB)',  data = df,ax=ax)
    plt.title(str(i)+' vs GSNRSpan(dB)')
    ticks = plt.setp(ax.get_xticklabels(),rotation=90)
    plt.show()
```

```python
df.drop("GSNRSpan(dB)", axis=1).apply(lambda x: x.corr(df['GSNRSpan(dB)'])) # to calculate correlation between each feature and the target 
```

```python
# to visualize the corralation heatmap is plotted
fig,ax = plt.subplots(figsize=(8,6))
correlation = df.select_dtypes(include=['float64','int64']).iloc[:,:].corr()
sns.heatmap(correlation,ax=ax,vmax=0.2,vmin=-0.7,square=True,cmap='viridis', linewidths=2)
plt.title('Numeric Feature heatmap')
```
